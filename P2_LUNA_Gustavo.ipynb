## Determinar la categoría de producto más vendida.

```python
# Con esta funcion podemos crear una tabla que nos muestre la cantidad de valores y su porcentaje
def tipos_porcentaje(df):
    # Calcular la cantidad de valores por columna
    valores =df.value_counts()
    
    total_valores=len(df)
    
    # Calcular el porcentaje de valores por columna
    porcentaje_valores = round(valores / total_valores * 100, 2)
    
    # Crear un DataFrame para almacenar los resultados
    tabla_valores = pd.DataFrame({'Cantidad': valores, 'Porcentaje': porcentaje_valores})
    
    #print(tabla_valores)
    
    return tabla_valores
```

```python
categorias=tipos_porcentaje(df_resultante.product_category_name)
categorias.reset_index(inplace=True)
categorias
```

```python
categoria_mas_vendida=categorias.product_category_name.loc[0]
cantidad_categoria_mas_vendida=categorias.Cantidad.loc[0]
print(separador)
print(f"La categoría de producto mas vendida con un total de {cantidad_categoria_mas_vendida} unidades fue: '{categoria_mas_vendida}'")
print(separador)
```
